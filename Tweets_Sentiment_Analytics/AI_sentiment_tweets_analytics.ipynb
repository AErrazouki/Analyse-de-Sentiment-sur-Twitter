# Scraping avec ntscraper

```python
from ntscraper import Nitter
```

```python
scraper = Nitter()
```

# La thematique est l'inteligence artificiel (AI)

```python
tweets = scraper.get_tweets("AI", mode = 'hashtag', number=1000)
```

# L'affichage des tweets format json

```python
tweets
```

# Afficher les donnée en DataFrame

```python
import pandas as pd
final_tweets = []
for x in tweets['tweets']:
    data = [x['link'], x['text'],x['date'],x['stats']['likes'],x['stats']['comments']]
    final_tweets.append(data)
```

```python
dat = pd.DataFrame(final_tweets, columns =['twitter_link','text','date','likes','comments'])
```

```python
dat
```

# Transformer au fichier CSV

```python
dat.to_csv("ChatGPT_data.csv")
```
